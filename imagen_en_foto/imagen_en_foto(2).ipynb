```python
from PIL import Image
from numpy import array
import numpy as np

#Se abre la imagen.
foto=Image.open("ho.jpeg");
#Se redimenciona pa mas facil.
ancho = 240
alto = 320
foto=foto.resize((ancho,alto))
#foto.show()
#Se colocan marcadores en los extremos de la hoja.
for i in range (foto.size[0]): 
    for j in range (foto.size[1]):
        if i==44 and j==212:
            foto.putpixel((i,j),(255,0,0))
        if i==175 and j==216:
            foto.putpixel((i,j),(255,0,0))
        if i==159 and j==79:
            foto.putpixel((i,j),(255,0,0))
        if i==73 and j==76:
            foto.putpixel((i,j),(255,0,0))
#foto.show()
#Coordenadas de la hoja.
p1=44,212
p2=175,216
p3=159,79
p4=73,76
#Coordenadas del pan.
q1 = 0,0
q2 = 50,0
q3 = 0,50
q4 = 50,50
#Se abre una nueva imagen.
vale=Image.open("pan.jpeg")
#Se redimenciona para que tenga un area similar a la hoja.
ancho2 = 122
alto2 = 121
vale=vale.resize((ancho2,alto2))
#Se hace una matriz de ayuda
c33=1

c11=int(1.080*c33)
c12=int(-2.372*c33)
c13=int(44*c33)
c21=int(0.040*c33)
c22=int(-2.75*c33)
c23=int(212*c33)
c31=int(0)
c32=int(0.021*c33)

Cs=[[c11, c12, c13],
          [c21, c22, c23],
          [c31, c32, c33]]

#print(Cs)
#Se hace una nueva foto
foto2=foto
#Se asignan pixeles
for x in range(vale.size[0]):
    for y in range(vale.size[1]):
        x2=(c11*x+c12*y+c13*1)
        y2=(c21*x+c22*y+c23*1)
        z2=(c31*x+c32*y+c33*1)

        x3=int(abs(x2/z2))
        y3=int(abs(y2/z2))

        #print(x3,y3)
        if x3<vale.size[0] and y3<vale.size[1]:
            r=vale.getpixel((x3, y3))[0]
            g=vale.getpixel((x3, y3))[1]
            b=vale.getpixel((x3, y3))[2]
        
            foto2.putpixel((x, y), (r, g, b))
foto2.show()
```
