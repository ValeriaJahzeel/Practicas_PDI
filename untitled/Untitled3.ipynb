```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# Funcion para convertir a imagen rgb


def img_rgb(imagen, wid, hght, rgb):
    newImage = np.empty((hght, wid), np.uint8)
    for i in range(hght):
        for j in range(wid):
            pixel = imagen[i][j]
            if rgb == "rojo":
                newImage[i][j] = pixel[0]
            elif rgb == "verde":
                newImage[i][j] = pixel[1]
            elif rgb == "azul":
                newImage[i][j] = pixel[2]
    return newImage

# Funcion para obtener el kernel cubico


def kernel_cub(x):
    val = 0
    a = 0.5

    if (0 <= np.abs(x)) and (np.abs(x) < 1):
        val = np.abs(x)*3 - 2*np.abs(x)*2 + 1.0
    elif (1 <= np.abs(x)) and (np.abs(x) < 2):
        val = -np.abs(x)*3 + 5*np.abs(x)*2 - 8*np.abs(x) + 4.0
    else:
        val = 0
    return val


def interpolation(data, p_ij):
    q = 0
    for j in range(0, 3):
        # Obtener las coordenadas
        v = int(np.floor(p_ij[1] - 1 + j))
        p = 0
        for i in range(0, 3):
            # Obtener las coordenadas
            u = int(np.floor(p_ij[0]) - 1 + i)
            # Obtener el valor del pixel interpolado
            p = p + data[u, v] * kernel_cub(p_ij[0] - u)
        # Accumular el valor del pixel
        q = q + p * kernel_cub(p_ij[1] - v)
    return q


def bicubic_interpol(data, by_x, by_y):
    # Inicializar imagen nueva
    new_data = np.zeros((int(data.shape[0]*by_x), int(data.shape[1]*by_y)))

# Iterar la imagen primero columnas y luego filas
    for i in range(0, new_data.shape[0] - int(by_x)):
        for j in range(0, new_data.shape[1] - int(by_y)):
            # Se guarda la informacion de la interpolacion del pixel actual
            new_data[i, j] = interpolation(
                data, (float(i)/by_x, float(j)/by_y))

    # Se regresa la nueva imagen
    return new_data[:int((data.shape[0]-1)*by_x), :int((data.shape[1]-1)*by_y)]


 # Imagen
img = Image.open("px.jpg")
img_aux = np.copy(img)
wid, hght = img.size
print("Image size: " + str(wid) + "x" + str(hght))
img_green = img_rgb(img_aux, wid, hght, "verde")
plt.imshow(img_green)


# Interpolacion cubica
new_img_bicubic = bicubic_interpol(img_green, 5.0, 5.0)
plt.imshow(new_img_bicubic)
```
