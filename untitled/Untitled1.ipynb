```python
"""
21-03-2023
Valeria Castaon
Programa que realiza una transformacion de color, pide el valor en RGB y el punto central recolorea la imagen. 
 Usar gris precisa
"""
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# pide los valores
vr = int(input("Ingrese el valor de R: "))
vg = int(input("Ingrese el valor de G: "))
vb = int(input("Ingrese el valor de B: "))   
central = int(input("Ingrese el valor central: "))
    
img = Image.open("imagenes_transformacion/BB.jpg")  #lee la imagen
matriz = np.array(img)  #convierte la imagen en un arreglo
filas,columnas,dimensiones = matriz.shape # filas, columnas y dimensiones de la matriz

#separa la imagen en sus 3 colores rgb
r, g, b = img.split()

#obtiene la matriz de colores
matriz_rojo = np.array(r)
matriz_verde = np.array(g)
matriz_azul = np.array(b)
    
#plt.imshow(img)


#cpnvierte la imagen a grises
#https://www.delftstack.com/es/howto/python/convert-image-to-grayscale-python/

img = mpimg.imread('imagenes_transformacion/BB.jpg')
R, G, B = img[:,:,0], img[:,:,1], img[:,:,2]
gris = 0.21 * R + 0.72 * G + 0.07 * B
plt.imshow(gris)
#plt.show()


# realiza la transformada de la matriz
rt = np.zeros((filas,columnas))
gt = np.zeros((filas,columnas))
bt = np.zeros((filas,columnas))

for i in range(filas):
    for j in range(columnas):
        if gris[i,j] < central:
            rt[i,j] = vr*gris[i,j]/128
            gt[i,j] = vg*gris[i,j]/128
            bt[i,j] = vb*gris[i,j]/128
        else:
            rt[i,j] = vr + ((gris[i,j]-central)/(255-central))
            gt[i,j] = vg + (gris[i,j]-central)/(255-central)
            bt[i,j] = vb + (gris[i,j]-central)/(255-central)
            

# matriz que contiene la nueva imagen 
union_imgs = np.zeros((filas,columnas,3), dtype=np.uint8)
        
# Asigna valores a cada canal de color
union_imgs[:,:,0] = rt  # valores de rojo
union_imgs[:,:,1] = gt  # valores de verde
union_imgs[:,:,2] = bt  # valores de azul

# Crea una imagen RGB desde la matriz tridimensional
imagen_rgb = Image.fromarray(union_imgs)

# Guarda la imagen en un archivo
imagen_rgb.save("imagenes_transformacion/transformadaBB.jpg")
print("Imagen lista")
plt.imshow(imagen_rgb)

```
