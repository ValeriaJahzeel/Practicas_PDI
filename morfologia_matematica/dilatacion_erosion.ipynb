```python
#def con_and():
#Se suben imagenes
yo=Image.open("Pose.jpg")
fondo=Image.open("Fondo.jpg")
#Se cambian de tamaÃ±o
ancho=320
alto=240
yo=yo.resize((ancho, alto))
fondo=fondo.resize((ancho, alto))
#Se sacan canales rgb
#yo.show()
#Fondo
fondo_rojo=fondo.getchannel("R")
fondo_verde=fondo.getchannel("G")
fondo_azul=fondo.getchannel("B")
#Yo
yo_rojo=yo.getchannel("R")
yo_verde=yo.getchannel("G")
yo_azul=yo.getchannel("B")
#Se hace una nueva imagen
tercera=fondo
#Se le sancan arreglos
ar_f_r=array(fondo_rojo)
ar_f_g=array(fondo_verde)
ar_f_b=array(fondo_azul)

ar_y_r=array(yo_rojo)
ar_y_g=array(yo_verde)
ar_y_b=array(yo_azul)
```

```python
#Se saca el absoluto
d_rojo = abs()
for x in range (yo.size[0]):
    for y in range (yo.size[1]):
        abs_r=abs(ar_f_r[y][x]-ar_y_r[y][x])
        abs_g=abs(ar_f_g[y][x]-ar_y_g[y][x])
        abs_b=abs(ar_f_b[y][x]-ar_y_b[y][x])
        #Se le pasan a la imagen
        tercera.putpixel((x,y),(abs_r, abs_g, abs_b))

#tercera.show()

#Se pasa a grises
gris=tercera.convert("L")
ar_t_g=array(gris)

#Se umbraliza xd
for x in range(yo.size[0]):
    for y in range(yo.size[1]):
        if ar_t_g[y][x]<180:
            tercera.putpixel((x,y),(255,255,255))
        else:
            tercera.putpixel((x,y),(0,0,0))
tercera.show()
```
